# Find Experiments With DAS Output

Set `date_folder` below, then run the notebook. It scans `Processed_data/Audio/<date_folder>/<exp>` and reports which experiments already contain the requested DAS output folder.

```python
from pathlib import Path

import pandas as pd

BASE_AUDIO = Path("/mnt/home/neurostatslab/ceph/saneslab_data/gily_data/Processed_data/Audio")
DAS_RELATIVE_PATH = Path("calls_confident/entropy_thr_default_0.30_hf_warble_0.60")
```

```python
date_folder = "2026_02"
```

```python
month_root = BASE_AUDIO / date_folder
if not month_root.exists():
    raise FileNotFoundError(f"Month folder not found: {month_root}")

records = []
for exp_dir in sorted(path for path in month_root.iterdir() if path.is_dir() and path.name.isdigit()):
    das_dir = exp_dir / DAS_RELATIVE_PATH
    records.append({
        "exp": int(exp_dir.name),
        "exp_path": str(exp_dir),
        "das_run": das_dir.exists(),
        "das_path": str(das_dir),
    })

results_df = pd.DataFrame(records).sort_values("exp").reset_index(drop=True)
done_df = results_df[results_df["das_run"]].reset_index(drop=True)
missing_df = results_df[~results_df["das_run"]].reset_index(drop=True)

print(f"Month folder: {month_root}")
print(f"Experiments found: {len(results_df)}")
print(f"DAS already run: {len(done_df)}")
print(f"DAS missing: {len(missing_df)}")
```
